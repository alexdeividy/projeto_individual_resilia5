# Relatório analítico sobre a evasão de funcionários utilizando as pesquisas feitas pela empresa e pela HRBP

- A pedido do Chief Marketing Officer(CMO), estarei realizando este relatório com o objetivo de compreender os motivos que estão fazendo os funcionários da empresa pedirem demissão. O foco principal é entender e analisar os números, e tomar uma conclusão sobre o que está motivando os funcionários a saírem da empresa, trazer insights para melhorar o entendimento, e buscar soluções para tais problemas, melhorar a relação entre empresa-funcionário e diminuir o número de evasões.

- Começaremos então resolvendo algumas questões que foram predefinidas e analisaremos os dados fornecidos pela empresa por meio de uma pesquisa após o pedido de demissão, e também pela pesquisa feita pela HRBP. Realizaremos novos gráficos para visualizar melhor os dados e facilitar o entendimento das informações aqui obtidas.

razões para saída.png

- O gráfico funciona por meio de localização no quadro, sendo Y para o Impacto(peso) e X para a frequência que o Motivo aparece.

- O gráfico é formado utilizando o modelo de círculos coloridos que fica maior de acordo com o número de vezes que o Motivo é contabilizado, onde cada um dos círculos representa um Motivo, sendo divididos em quatro localizações no quadro, de acordo com o Impacto(peso) e Frequência do Motivo.

- Nos gráficos, os círculos são divididos da seguinte forma: pro lado direito, quanto mais frequente é o Motivo, e pra esquerda, menos frequênte(eixo X)
Para a parte de cima, maior é o Impacto(peso) do Motivo, quanto mais pra baixo, menor.

## 2. Que observações você pode fazer a partir desses dados? Coloque duas conclusões específicas em palavras. Escreva sentenças completas.

- Os dados nos mostram que os funcionários não se relacionam muito bem com a empresa, vários motivos poderiam ser evitados se houvesse uma colaboração maior entre os mesmos.

  - Reconhecimento: Outro motivo bem frequênte e preocupante é a Falta de Reconhecimento, tendo um impacto significativo nas pesquisas. Mostrando que os funcionários não se sentem reconhecidos pelo trabalho feito. Isso desmotiva ainda mais e leva a perda de desempenho ou até mesmo ao pedido de demissão. Seria legal criar alguns costumes como dar Feedbacks, premiações e incentivos, isso ajudaria muito no desempenho e ânimo dos funcionários.

## 3. O que não é ideal ou poderia ser evitado na visualização em sua forma atual? Que perguntas você faria ou que feedback daria para quem fez o gráfico?

- Analisando o gráfico podemos ver alguns pontos que podem ser alterados, e com isso resultar em uma visualização mais limpa e intuitiva.

  - O tamanho dos círculos dificulta a leitura as vezes, primeiro na diferenciação de alguns números parecidos, segundo que em Motivos com poucos números o círculo fica pequeno demais e terceiro, com pouca variação de cores, alguns círculos ficam de difícil diferenciação por ter cores parecidas

  - A forma em que o gráfico foi elaborado não é a melhor possível, a legenda é quase maior que o gráfico em si e chama tanta atenção que dissipa um pouco o foco do gráfico. Falta uma identificação de valores nos eixos para melhor identificação, pois sem os indicativos de números fica bastante confuso saber onde cada círculo se encontra na tabela de Impacto e Frequência.

- Um gráfico tem que ser de fácil entendimento, muitos elementos e cores parecidas geram extrema confusão na hora da visualização dos gráficos.

- Sendo assim, perguntaria ao colega alguns dos motivos pelos quais ele utilizou tal estrutura, para tentar entender seu pensamento e assim alterar o projeto em colaboração com o mesmo. Sugerindo melhorias juntos e refletindo sobre as escolhas feitas. Colocando atenção em certos pontos como o fácil entendimento de quem for olhar o gráfico, precisamos de uma estrutura limpa e clara, sem muitos elementos nem cores demasiadas nem parecidas demais neste caso em particular.


## 4. Crie três maneiras diferentes de mostrar esses dados e liste quais são algumas vantagens e insuficiências de cada uma.

- Para um melhor entendimento dos dados em relação ao cálculo que precisa ser feito dos números que os motivos aparecem, pelo impacto que eles causam, será descrito como (peso) os números de impacto, e na maioria dos gráficos os números utilizados são dos dados já multiplicados pelo peso da categoria do Motivo.

### Comparação dos dados Lado a Lado com o **Gráfico de Barras Vertical**

COMPARAÇÃO EMPRESA_HRBP_PESO.png

- Pontos Fortes:

  - Podemos ver lado a lado os valores de ambas as pesquisas e o Peso(impacto) do Motivo.

  - Facilidade de comparação dos dados, separados em apenas 3 cores o gráfico fica muito limpo e bom de visualizar e entender

  - Por utilizar 3 barras para cada dado, se torna bastante compacto a comparação, sem a necessidade de criar um gráfico para cada dads

- Pontos Fracos:

  - A exatidão dos dados fica comprometida por ter 3 dados em conjunto, limitando a capacidade de mais detalhes

  - Mesmo sendo apenas 3 cores diferentes, ainda assim prejudica um pouco na visualização, pois muitas cores torna confuso o entendimento.

### **Gráfico de barras horizontal**

PESQUISA EMPRESAporIMPACTO.png

PESQUISA HRBPporIMPACTO.png

- Pontos Fortes:

  - Muito bom para comparações, a medida que a cor vai escurecendo e a barra aumentando é muito intuitiva e auto-explicativa sobre o gráfico.

  - Excelente para diferenciar os Motivos entre si.


- Pontos Fracos:

  - Mesmo tendo linhas indicativas e valores, alguns dados só são diferenciado pela posição, pois diferenciar o valor é quase impossível.

  - Para uma comparação com outras pesquisas, é preciso fazer um gráfico para cada uma, tornando meio difícil comparar as pesquisas entre si.


### **Tabela de Destaque**

VOLUME COM NÚMEROS.png

- Pontos Fortes:

  - Bastante compacto e preciso para difenciar Motivos com pequenas diferenças de números.

  - Por mostrar os números, a descrição e a força da cor, é bastante simples de compreender.


- Pontos Fracos:

  - A cor não fica muito parecida em valores muito próximos.

  - Necessita bastante atenção nos números, pois não diferencia os Motivos como no gráfico de barras por exemplo, todos estão alinhados.


## **Colunas Empilhadas**

VOLUME COM CORES DIFERENTES.png

- Pontos Fortes:

  - Fácil de diferenciar os Motivos por cada um ter uma cor diferente.

  - Fácil de entender qual Motivo está vencendo pois está em ordem de quantidade, da esquerda para a direita.


- Pontos Fracos:

  - Muitas cores pode causar confusão no leitor.

  - Mesmo tendo a legenda dos números logo abaixo, não conseguimos diferenciar tão bem em que número estamos.
